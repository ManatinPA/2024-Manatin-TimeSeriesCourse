```python
import stumpy
import numpy as np


data = np.loadtxt('datasets/Rotation Matrix.txt')
window_size = 15

# Вычисляем матричный профиль
matrix_profile = stumpy.stump(data, m=window_size)
profile_values = matrix_profile[:, 0]

# Устанавливаем порог (среднее значение матричного профиля)
threshold = np.mean(profile_values)


chains = []
current_chain = [0]  
for i in range(1, len(profile_values)):
    if profile_values[i] <= threshold: 
        current_chain.append(i)
    else:
        if len(current_chain) > 1:
            chains.append(current_chain)
        current_chain = [i]  

if len(current_chain) > 1:
    chains.append(current_chain)

# Выводим результат
print("Найденные цепочки:")
for chain in chains:
    print(f"Цепочка из {len(chain)} элементов: {chain}")
```
